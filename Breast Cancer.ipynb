```python
#feature scaling
#split our data set into independent and dependent data sets
#independent --> X
#dependent --> Y
X = df.iloc[:,2:31].values
Y = df.iloc[:,1].values
```

```python
#80:20 ratio
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y , test_size = 0.20 , random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train
```

```python
def models(X_train, Y_train):
  from sklearn.linear_model import LogisticRegression
  log = LogisticRegression(random_state = 0)
  log.fit(X_train, Y_train.astype(int))

  from sklearn.tree import DecisionTreeClassifier
  tree = DecisionTreeClassifier(criterion= 'entropy', random_state= 0)
  tree.fit(X_train, Y_train.astype(int))

  from sklearn.ensemble import RandomForestClassifier
  forest = RandomForestClassifier(n_estimators= 10, criterion= 'entropy',random_state= 0)
  forest.fit(X_train, Y_train.astype(int))

  #print the accuracy of each model on the training dataset
  print('The accuracy of Logistic Regression: ',log.score(X_train, Y_train.astype(int)))
  print('The accuracy of Decision Tree: ',tree.score(X_train, Y_train.astype(int)))
  print('The accuracy of Random Forest: ',forest.score(X_train, Y_train.astype(int)))

  return log, tree, forest



```

```python
model = models(X_train, Y_train)
```

```python
from sklearn.metrics import confusion_matrix
predictions = model[0].predict(X_test)
predictions = predictions.astype(int)
Y_test = Y_test.astype(int)

cm = confusion_matrix(Y_test, model[0].predict(X_test))
tp = cm[0][0]
tn = cm[1][1]
fn = cm[1][0]
fp =cm[0][1]
print(cm)
print('Accuracy: ',(tp+tn)/(tp+tn+fp+fn))
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

for i in range(len(model)):
  print('Model: ',i)
  print(classification_report(Y_test, model[i].predict(X_test)))
  print(accuracy_score(Y_test, model[i].predict(X_test)))
  print()
```
